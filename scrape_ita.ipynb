```python
def manage(sc_list, verbose=False):
    #df = data_processor.get_data_or_empty_dataframe()
    while True:
        dt_now = datetime.now()
        if dt_now < datetime(dt_now.year, dt_now.month, dt_now.day, 9, 0, 0):
            time.sleep(0.8)
            continue
        elif (datetime(dt_now.year, dt_now.month, dt_now.day, 11, 30, 0) <= dt_now and dt_now < datetime(dt_now.year, dt_now.month, dt_now.day, 12, 30, 0)):
            if verbose:
                print("Updated token")
            time.sleep(0.8)
            continue
        elif not (datetime(dt_now.year, dt_now.month, dt_now.day, 9, 0, 0) <= dt_now and dt_now < datetime(dt_now.year, dt_now.month, dt_now.day, 15, 0, 0)):
            break
        print(dt_now)

        # 毎度空にする
        df = pd.DataFrame()
        for sc in sc_list:
            try:
                res = kabucom.get_board(sc)
                tmp_df = convert_to_df(res)
                df = pd.concat([df, tmp_df])
            except Exception as e:
                print("Failed to gather : " + str(sc) + " " + str(dt_now) + str(e))
        
        # bulkで書き込み
        db.insert(df)  
        #save(df.drop_duplicates())
        time.sleep(0.8)

    if len(df) == 0:
        raise Exception("DataFrame is empty")

    return df

df = manage(sc_list=sc_list, verbose=True)
df
```
